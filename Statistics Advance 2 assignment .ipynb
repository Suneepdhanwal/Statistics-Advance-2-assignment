## **Therory Questions**

**Ques.1** What is hypothesis testing in statistics?

**Ans.1** Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It involves formulating a hypothesis, conducting an experiment or survey, and using statistical analysis to determine whether there is enough evidence to reject the null hypothesis.

**Ques.2** What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Ans.2** The null hypothesis (denoted as H0) is a statement that there is no effect or no difference, and it serves as the default assumption. The alternative hypothesis (denoted as H1 or Ha) is a statement that there is an effect or a difference. Hypothesis testing aims to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

**Ques.3** What is the significance level in hypothesis testing, and why is it important?

**Ans.3** The significance level (denoted as ð›¼) is the probability of rejecting the null hypothesis when it is true. It is set by the researcher and commonly used values are 0.05, 0.01, and 0.10. The significance level is important because it defines the threshold for determining whether the observed data is statistically significant.

**Ques.4** What does a P-value represent in hypothesis testing?

**Ans.4** A P-value is the probability of obtaining results at least as extreme as the observed results, assuming that the null hypothesis is true. It provides a measure of the strength of the evidence against the null hypothesis.

**Ques.5** How do you interpret the P-value in hypothesis testing?

**Ans.5** If the P-value is less than or equal to the significance level (ð›¼), we reject the null hypothesis and conclude that there is evidence to support the alternative hypothesis. If the P-value is greater than the significance level, we fail to reject the null hypothesis.

**Ques.6** What are Type 1 and Type 2 errors in hypothesis testing?

**Ans.6**  Type 1 Error (False Positive): Rejecting the null hypothesis when it is true.

Type 2 Error (False Negative): Failing to reject the null hypothesis when it is false.

**Ques.7** What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

**Ans.7**
One-Tailed Test: Tests for an effect in only one direction (either greater than or less than a certain value).

Two-Tailed Test: Tests for an effect in both directions (both greater than and less than a certain value).

**Ques.8** What is the Z-test, and when is it used in hypothesis testing?

**Ans.8** The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means when the population variance is known. It is used when the sample size is large (usually n>30).

**Ques.10** What is the T-distribution, and when should it be used instead of the normal distribution?

**Ans.10** The T-distribution is a probability distribution used when the sample size is small (usually n < 30) and the population variance is unknown. It is similar to the normal distribution but has heavier tails.

**Ques.11** What is the difference between a Z-test and a T-test?

**Ques.1** Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results ?

```python
import numpy as np
from scipy import stats

# Sample data
sample = [50, 52, 55, 49, 51, 54, 56, 48, 53, 50]
# Population mean
population_mean = 52
# Population standard deviation (assumed known)
population_std = 3

# Calculate sample mean
sample_mean = np.mean(sample)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
# Calculate P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

**Ques.5** Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_type_errors(alpha, beta, sample_size, population_mean, sample_mean):
    z_alpha = stats.norm.ppf(1 - alpha/2)
    z_beta = stats.norm.ppf(beta)

    plt.figure(figsize=(10, 6))

    # Plot the null hypothesis distribution
    x_null = np.linspace(population_mean - 4, population_mean + 4, 1000)
    y_null = stats.norm.pdf(x_null, population_mean, 1)
    plt.plot(x_null, y_null, label='Null Hypothesis Distribution')

    # Plot the alternative hypothesis distribution
    x_alt = np.linspace(sample_mean - 4, sample_mean + 4, 1000)
    y_alt = stats.norm.pdf(x_alt, sample_mean, 1)
    plt.plot(x_alt, y_alt, label='Alternative Hypothesis Distribution')

    # Highlight Type 1 error region
    plt.fill_between(x_null, 0, y_null, where=(x_null < population_mean - z_alpha) | (x_null > population_mean + z_alpha), color='red', alpha=0.5, label='Type 1 Error Region (Î±)')

    # Highlight Type 2 error region
    plt.fill_between(x_alt, 0, y_alt, where=(x_alt > sample_mean - z_beta) & (x_alt < sample_mean + z_beta), color='blue', alpha=0.5, label='Type 2 Error Region (Î²)')

    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors')
    plt.legend()
    plt.show()

# Example parameters
alpha = 0.05
beta = 0.2
sample_size = 30
population_mean = 50
sample_mean = 52

calculate_type_errors(alpha, beta, sample_size, population_mean, sample_mean)
```

**Ques.9** Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample, confidence_level=0.95):
    # Calculate sample mean and standard error
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(len(sample))

    # Calculate the confidence interval
    confidence_interval = stats.t.interval(confidence_level, len(sample)-1, loc=sample_mean, scale=standard_error)
    return confidence_interval

# Example usage
sample = [50, 52, 55, 49, 51, 54, 56, 48, 53, 50]
confidence_level = 0.95
confidence_interval = calculate_confidence_interval(sample, confidence_level)

print(f"Sample Mean: {np.mean(sample)}")
print(f"Confidence Interval ({confidence_level*100}%): {confidence_interval}")

#Confidence intervals are crucial in statistics as they provide a range within which we can be reasonably confident that the true population parameter lies.
```

**Ques.10* Write a Python program to calculate the margin of error for a given confidence level using sample data?

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(len(sample))
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=len(sample)-1)
    margin_of_error = critical_value * standard_error
    return margin_of_error

# Example usage
sample = [50, 52, 55, 49, 51, 54, 56, 48, 53, 50]
confidence_level = 0.95
margin_of_error = calculate_margin_of_error(sample, confidence_level)

print(f"Margin of Error ({confidence_level*100}%): {margin_of_error}")
```

**Ques.11** Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

```python
def bayesian_inference(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

# Example usage
prior = 0.01  # Prior probability (P(A))
likelihood = 0.9  # Likelihood (P(B|A))
evidence = 0.05  # Evidence (P(B))

posterior = bayesian_inference(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior}")

# Explanation:
# Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)
# P(A|B): Posterior probability
# P(B|A): Likelihood
# P(A): Prior probability
# P(B): Evidence
```

**Ques.16** Implement an F-test using Python to compare the variances of two random samples?

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f_stat = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    p_value = 1 - f.cdf(f_stat, df1, df2)
    return f_stat, p_value

# Sample data
sample1 = [10, 12, 14, 16, 18]
sample2 = [8, 9, 10, 11, 12]

# Perform F-test
f_stat, p_value = f_test(sample1, sample2)
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

**Ques.19** Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    # Check normality using Shapiro-Wilk test
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i} - Shapiro-Wilk Test: Stat = {stat}, P-value = {p}")

    # Check equal variance using Levene's test
    stat, p = levene(*groups)
    print(f"Levene's Test: Stat = {stat}, P-value = {p}")

    # Plot distributions
    for i, group in enumerate(groups, 1):
        sns.histplot(group, kde=True, label=f'Group {i}')
    plt.legend()
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('ANOVA Assumptions Check - Distributions')
    plt.show()

# Example usage
group1 = [50, 52, 55, 49, 51]
group2 = [60, 62, 65, 59, 61]
group3 = [70, 72, 75, 69, 71]

check_anova_assumptions([group1, group2, group3])
```

```python
import numpy as np
from scipy.stats import chi2

# Sample data
sample = [50, 52, 55, 49, 51, 54, 56, 48, 53, 50]
# Population variance (assumed known)
population_variance = 5

# Calculate sample variance
sample_variance = np.var(sample, ddof=1)
sample_size = len(sample)
chi2_stat = (sample_size - 1) * sample_variance / population_variance

# Calculate P-value
p_value = 1 - chi2.cdf(chi2_stat, df=sample_size - 1)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample variance is significantly different from the population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the population variance.")
```

**Ques.26** Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?

```python
import numpy as np

# Generate two random datasets
np.random.seed(42)
data1 = np.random.normal(10, 2, 100)  # mean = 10, std = 2, size = 100
data2 = np.random.normal(12, 3, 100)  # mean = 12, std = 3, size = 100
```

**Ques.27** Perform a Chi-square test for goodness of fit with simulated data and analyze the results?

```python
import scipy.stats as stats

# Calculate variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Calculate F-statistic
F = var1 / var2

# Get the degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Calculate the p-value
p_value = stats.f.cdf(F, df1, df2) if F < 1 else 1 - stats.f.cdf(F, df1, df2)

print(f"F-statistic: {F}, p-value: {p_value}")
```
